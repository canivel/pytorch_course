```python
import numpy as np
import torch
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
torch.manual_seed(1)
X = torch.randn(100, 1)*3.5
y = X + 4.5*torch.randn(100, 1)
```

```python
plt.plot(X.numpy(), y.numpy(), 'o')
plt.xlabel('x')
plt.ylabel('y')
```

```python
class LR(torch.nn.Module):
    
    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear = torch.nn.Linear(input_size, output_size)
    
    def forward(self, x):
        pred = self.linear(x)
        return pred
```

```python
model = LR(1, 1)
print(model)
```

```python
[w, b] = model.parameters()
def get_params():
    return (w[0][0].item(), b[0].item())
```

```python
def fit(plot=True, title="Fit"):
    w1, b1 = get_params()
    x1 = np.array([-13, 10])
    y1 = w1*x1+b1
    if plot:
        plt.title(title)
        plt.plot(x1, y1, 'r')
        plt.scatter(X, y)
```

```python
fit()
```
